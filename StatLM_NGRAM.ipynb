```python
import re

def cria_ngrama(texto, n):
    texto = texto.lower()
    texto = re.sub(r'[^a-zA-Z0-9\s]', '', texto)
    palavras = texto.split()
    ngrams = []
    for i in range(len(palavras) - n + 1):
      ngram = tuple(palavras[i:i+n])
      ngrams.append(ngram)
    return ngrams
```

```python
cria_ngrama('ola mundo, eu sou feliz', 2)
```

```python
n=2
setores = {}
ngramas_setor = {}
```

```python
for reclamacao, setor in dados.items():
  if setor not in setores:
      setores[setor] = 0
      ngramas_setor[setor] = {}
  setores[setor] += 1
  for ngram in cria_ngrama(reclamacao,n):
    if ngram not in ngramas_setor[setor]:
      ngramas_setor[setor][ngram] = 0
    ngramas_setor[setor][ngram] += 1
```

```python
reclamacao_teste = 'O produto no est igual ao anunciado'
ngrams = cria_ngrama(reclamacao_teste,n)
max_prob = float('-inf')
setor_predicao = None

for label in setores.keys():
    print('testando para o label >> ',label)
    prob = 0
    total_ngramas_setor = sum(ngramas_setor[label].values())
    for ngram in ngrams:
      total_ngramas_encontrados = ngramas_setor[label].get(ngram, 0)
      prob += total_ngramas_encontrados / total_ngramas_setor
      if prob > max_prob:
        max_prob = prob
        setor_predicao = label
    print('maior probabilidade >> ',prob)
```
